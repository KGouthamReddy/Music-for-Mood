<h2>Trail 4 : Marking the position and emotion on the picture</h2>

```python
from deepface import DeepFace
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('goutham2.jpg')
```

```python
plt.imshow(img[:,:,::-1])
plt.title('img')
plt.show()
```

```python
predictions = DeepFace.analyze(img, actions = ['emotion'])
```

```python
print(predictions)
```

```python
print(predictions['dominant_emotion'])
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #print(faceCascade.empty())
faces = faceCascade.detectMultiScale(gray,1.1,4)
    
    #Draw a rectangle around the faces
for(x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 3)
```

```python
plt.imshow(img[:,:,::-1])
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

    # Use putText() method for inserting text on video/image
cv2.putText(img, predictions['dominant_emotion'], (0, 200), font, 3, (0, 0, 255), 3, cv2.LINE_4);
                 
```

```python
plt.imshow(img[:,:,::-1])
```
